## Step 1: Vehicle Superclass
We will create a superclass `Vehicle` with a single attribute for vehicle type.


```python
# Vehicle superclass
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type
```

## Step 2: Automobile Subclass
The `Automobile` class inherits from `Vehicle` and adds attributes for year, make, model, doors, and roof type.


```python
# Automobile subclass
class Automobile(Vehicle):
    def __init__(self, year, make, model, doors, roof):
        super().__init__("car")  # Vehicle type is always "car" for this subclass
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof
```
