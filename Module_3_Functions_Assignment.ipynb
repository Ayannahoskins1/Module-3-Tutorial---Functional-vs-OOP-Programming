## 7.11 – Jump Rope Rhyme Maker  
Use the provided lists and print two-line rhymes.


```python
start1 = ["fee", "fie", "foe"]
rhymes = [
    ("flop", "get a mop"),
    ("fope", "turn the rope"),
    ("fa", "get your ma"),
    ("fudge", "call the judge"),
    ("fat", "pet the cat"),
    ("fog", "walk the dog"),
    ("fun", "say we're done"),
]
start2 = "Someone better"

for first, second in rhymes:
    line1 = ""
    for word in start1:
        line1 += word.capitalize() + "! "
    line1 += first.capitalize() + "!"

    line2 = start2 + " " + second + "."

    print(line1)
    print(line2)
    print()
```
