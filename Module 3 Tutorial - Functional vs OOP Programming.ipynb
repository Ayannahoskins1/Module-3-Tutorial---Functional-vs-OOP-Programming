## Object-Oriented Programming Exercise
**Sort an Array of 0s, 1s, and 2s**  
This exercise demonstrates OOP design using a class and method.


````python
# OOP Exercise - Sort 0s, 1s, and 2s

class Solution:
    def sort012(self, arr, n):
        low = 0
        mid = 0
        high = n - 1
        
        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1
        
        return arr

# Example run
arr = [0, 2, 1, 2, 0, 1]
sol = Solution()
sol.sort012(arr, len(arr))
````

## Functional Programming Exercise
**Binary Search Using Lambda Expressions**  
This exercise demonstrates functional-style programming using lambda functions.


````python
# Functional Programming Exercise - Binary Search

def binarysearch(arr, k):
    left, right = 0, len(arr) - 1
    ans = -1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == k:
            ans = mid          # store index
            right = mid - 1    # continue searching left for smallest index
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1

    return ans

# Example run
binarysearch([1, 2, 3, 4, 5], 4)   # Output: 3
binarysearch([11, 22, 33, 44, 55], 445)  # Output: -1
binarysearch([1, 1, 1, 1, 2], 1)   # Output: 0
````
